```python
def trend_line(X, Y):
    z = polyfit(X, Y, 1)
    p = poly1d(z)
    return p
```

```python
with open('enemies.txt', 'r') as enemies:
    for enemy in enemies:
        enemy = enemy[:-1]
        sgb = 'gb_results/' + enemy + '.txt'
        Ygb = genfromtxt(sgb)
        Xgb = [i for i in range(0, len(Ygb))]
        ssb = 'sb_results/' + enemy + '.txt'
        Ysb = genfromtxt(ssb)
        Xsb = [i for i in range(0, len(Ysb))]
        sdb = 'db_results/' + enemy + '.txt'
        Ydb = genfromtxt(sdb)
        Xdb = [i for i in range(0, len(Ydb))]
        
        figure(figsize=(24, 24))

        subplot(4, 3, 1)
        scatter(Xgb, Ygb[:, 0], marker='x', label='GB')
        p = trend_line(Xgb, Ygb[:, 0])
        plot(Xgb, p(Xgb), 'b--')
        scatter(Xsb, Ysb[:, 0], label='SB')
        p = trend_line(Xsb, Ysb[:, 0])
        plot(Xsb, p(Xsb), 'r--')
        scatter(Xdb, Ydb[:, 0], label='DB')
        p = trend_line(Xdb, Ydb[:, 0])
        plot(Xdb, p(Xdb), 'g--')
        legend(loc='upper right')
        name = enemy + ': Post-Fight Health'
        title(name)
        xlabel('Iteration #')

        subplot(4, 3, 2)
        scatter(Xgb, Ygb[:, 1], marker='x', label='GB')
        p = trend_line(Xgb, Ygb[:, 1])
        plot(Xgb, p(Xgb), 'b--')
        scatter(Xsb, Ysb[:, 1], label='SB')
        p = trend_line(Xsb, Ysb[:, 1])
        plot(Xsb, p(Xsb), 'r--')
        scatter(Xdb, Ydb[:, 1], label='DB')
        p = trend_line(Xdb, Ydb[:, 1])
        plot(Xdb, p(Xdb), 'g--')
        legend(loc='upper right')
        name = enemy + ': Time to Kill'
        title(name)
        xlabel('Iteration #')
        figure()
```
