```python
%pylab inline
import pandas as pd
```

```python
def trend_line(X, Y):
    z = polyfit(X, Y, 1)
    p = poly1d(z)
    return p
dumbBotResults = "results_DB_optimal1.txt"
generalBotResults = "results_GB_optimal1.txt"
specialBotResults = "results_SB_optimal3.txt"
columns = ["enemy", "health", "time", "kill"]
dbDF = pd.read_csv(dumbBotResults, names=columns)
gbDF = pd.read_csv(generalBotResults, names=columns)
sbDF = pd.read_csv(specialBotResults, names=columns)

    
```

```python
for enemy in dbDF['enemy'].unique():
    Ygb = gbDF[gbDF['enemy'] == enemy]
    Xgb = [i for i in range(0, len(Ygb))]
    Ysb = sbDF[sbDF['enemy'] == enemy]
    Xsb = [i for i in range(0, len(Ysb))]
    Ydb = dbDF[dbDF['enemy'] == enemy]
    Xdb = [i for i in range(0, len(Ydb))]

    figure(figsize=(24, 24))

    ax = subplot(4, 3, 1)
    gbHealth = list(Ygb['health'])
    sbHealth = list(Ysb['health'])
    dbHealth = list(Ydb['health'])
    
    scatter(Xgb, gbHealth, marker='x', color='b', label='GB')
    p = mean(gbHealth)
    plot(Xgb, [p for i in Xgb], 'b--')
    scatter(Xsb, sbHealth, color='r', marker='+', label='SB')
    p = mean(sbHealth)
    plot(Xsb, [p for i in Xsb], 'r--')
    scatter(Xdb, dbHealth, color='g', marker='d', label='DB')
    p = mean(dbHealth)
    plot(Xdb, [p for i in Xdb], 'g--')
    legend(loc='upper right')
    name = enemy + ': Post-Fight Health'
    title(name)
    ax.set_ylim([-.5, 22])
    xlabel('Iteration #')

    ax = subplot(4, 3, 2)
    gbTime = list(Ygb['time'])
    sbTime = list(Ysb['time'])
    dbTime = list(Ydb['time'])
    scatter(Xgb, gbTime, marker='x', color='b', label='GB')
    u = mean(gbTime)
    plot(Xgb, [u for i in Xgb], 'b--')
    scatter(Xsb, sbTime, color='r', marker='+', label='SB')
    p = mean(sbTime)
    plot(Xsb, [p for i in Xsb], 'r--')
    scatter(Xdb, dbTime, color='g', marker='d', label='DB')
    p = mean(dbTime)
    plot(Xdb, [p for i in Xdb], 'g--')
    legend(loc='upper right')
    name = enemy + ': Time to Kill'
    title(name)
    ax.set_ylim([-.5, 27])
    xlabel('Iteration #')
    figure()
```
